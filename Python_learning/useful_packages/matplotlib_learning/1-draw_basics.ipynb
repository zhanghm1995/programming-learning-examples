## subplots

```python
# fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(6, 2), 
#                        sharey=True,sharex=True)
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(6, 2))
## set the background color of the figure
patch = fig.patch
patch.set_color("blue")

print(type(fig))
print(type(ax), ax.shape, type(ax[0, 0]))

for i in range(2):
    for j in range(3):
        ax[i, j].imshow(np.random.random((100, 200)))
        ax[i, j].axis('off')

        # set the background color of the subplot
        patch = ax[i, j].patch
        patch.set_color("green")

# fig.tight_layout()
plt.subplots_adjust(left=0, wspace=0, hspace=0.1)
plt.savefig("subplots.png")
```
