## mmcv

```python
import mmcv
import cv2
import numpy as np

print(mmcv.image.io.imread_backend)  # opencv2 by default

img_path = "data/obama.jpg"

img = mmcv.imread(img_path, "unchanged")

print(img.shape, img.dtype, img.min(), img.max())

img_cv2 = cv2.imread(img_path, cv2.IMREAD_UNCHANGED)
print(img_cv2.shape, img_cv2.dtype, img_cv2.min(), img_cv2.max())

diff = np.abs(img - img_cv2)
print(diff.max())

## PIL image
from PIL import Image
img_pil = Image.open(img_path)
img_pil = np.array(img_pil)
print(img_pil.shape, img_pil.dtype, img_pil.min(), img_pil.max())

diff = np.abs(img - img_pil)
print(diff.max())

img_pil_bgr = img_pil[..., ::-1]
diff = np.abs(img - img_pil_bgr)
print(diff.max())
```
